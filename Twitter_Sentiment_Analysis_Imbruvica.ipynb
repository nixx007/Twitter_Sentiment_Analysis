```python
import pandas as pd

# Merging all the tweets csv into one Dataframe.
df = pd.concat( 
    map(pd.read_csv, ['twitter_data_new.csv', 'twitter_data_01.csv', 'twitter_data_02.csv', 'twitter_data_03.csv', 'twitter_data_04.csv']), ignore_index=True) 
print(df) 
```

```python
#Removing Duplicate Rows

df.sort_values("Tweets", inplace=True) 
df.drop_duplicates(subset="Tweets", 
                     keep='first', inplace=True)
df.head()
```

```python
df.Replys = (df.Replys.replace(r'[KM]+$', '', regex=True).astype(float) * df.Replys.str.extract(r'[\d\.]+([KM]+)', expand=False).fillna(1).replace(['K','M'], [10**3, 10**6]).astype(int))
df.reTweets = (df.reTweets.replace(r'[KM]+$', '', regex=True).astype(float) * df.reTweets.str.extract(r'[\d\.]+([KM]+)', expand=False).fillna(1).replace(['K','M'], [10**3, 10**6]).astype(int))
df.Likes = (df.Likes.replace(r'[KM]+$', '', regex=True).astype(float) * df.Likes.str.extract(r'[\d\.]+([KM]+)', expand=False).fillna(1).replace(['K','M'], [10**3, 10**6]).astype(int))
df.fillna(0, inplace=True)
df.Likes
```

```python
# Fetching all physician tweets and their engagement rates.

physician_user = [tweet for tweet in df.UserTags if "Dr." in tweet or "Doctor" in tweet or "doctor" in tweet or "PhD" in tweet or "MD" in tweet or "MBBS" in tweet or "dr" in tweet]
# physician_user = [tweet for tweet in df.UserTags if if any(x in a_string for x in tweet)]
physician_user
engagement_rate = 0
physician_data = {}
physician_tweets = []
for i in physician_user:
    physician_tweets.append(df.loc[df['UserTags'] == i, 'Tweets'].item())
    engagement_rate = df.loc[df['UserTags'] == i, 'Likes'].item() + df.loc[df['UserTags'] == i, 'Replys'].item() + df.loc[df['UserTags'] == i, 'reTweets'].item()
    physician_data[i] = engagement_rate
physician_tweets
physician_data
```

```python
!pip install --user textblob

from textblob import TextBlob
```

```python
# Calculate sentiment analysis for physician tweets

positive_count, negative_count, neutral_count = 0, 0, 0
for tweet in physician_tweets:
    analysis = TextBlob(tweet)
    if analysis.sentiment.polarity > 0:
        positive_count += 1
    elif analysis.sentiment.polarity < 0:
        negative_count += 1
    else:
        neutral_count += 1

total_physician_tweets = len(physician_tweets)
positive_percentage = (positive_count / total_physician_tweets) * 100
negative_percentage = (negative_count / total_physician_tweets) * 100
neutral_percentage = (neutral_count / total_physician_tweets) * 100
print(positive_percentage, negative_percentage, neutral_percentage)
```

```python
KOL_list = []
engagement_metric = 0
for i in df.UserTags:
    try:
        engagement_metric = df.loc[df['UserTags'] == i, 'Likes'].item()
    except ValueError:
        engagement_metric = 0
    KOL_list.append({
        'User': i,
        'Engagement Metrics': engagement_metric
        })
sorted_KOL_list = sorted(KOL_list, key=lambda x: (x["User"], x["Engagement Metrics"]), reverse=True)
```

```python
# Displaying Results

print("Physician Sentiment Analysis:")
print(f"Positive Tweets: {positive_percentage}%")
print(f"Negative Tweets: {negative_percentage}%")
print(f"Neutral Tweets: {neutral_percentage}%")

print("\nEngagement Rate for Physician Tweets:")
print(f"Total Engagement Rate: {engagement_rate}")

print("\nKey Opinion Leaders (KOLs):")
for kol in sorted_KOL_list:
    print(f"Username: {kol['User']}, Engagement Metrics: {kol['Engagement Metrics']}")
```
